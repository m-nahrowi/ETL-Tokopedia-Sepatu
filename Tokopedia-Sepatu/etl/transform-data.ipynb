```python
# normalisasi data sold
def clean_text(text):
    if isinstance(text, str):
        # Hapus karakter non-alfanumerik kecuali koma, titik, dan spasi
        text = re.sub(r'[^a-zA-Z0-9,. ]', '', text)
        # Ubah koma (,) menjadi titik (.) untuk format numerik yang konsisten
        text = text.replace(',', '.')
        # Hapus kata-kata yang tidak diperlukan
        text = text.lower()
        text = text.replace('diskon', '').replace('cashback', '').replace('terjual', '').replace('terlaris', '').replace('preorder', '').replace('grosir', '').replace('ramadan ekstra', '').replace('sisa', '').replace('tokopedia fashion','').replace('beli lokal', '').replace('produk tearu','').replace(' ','').replace('produkterbaru','')
        # text = re.sub(r'(\d+)rb$', lambda x: str((int(x.group(1)) * 1000)), text)
        text = text if text.strip() else np.nan
        # text = re.sub(r'(\d+)rb', lambda x: str(int(x.group(0)) * 1000), text)
        
    return text
```

```python
df['sold'] = df['sold'].apply(clean_text)
```

```python
# for i in df['sold']:
#     print(i)
df.isnull().sum()
```

```python
def convert_value(x):
    if pd.isna(x):
        return x
    elif '.' in x:
        value = (float(x.replace('rb', '').replace('.', '')) * 1000)/10
    else:
        value = int(x.replace('rb', '')) * 1000
    return value

df['sold'] = df['sold'].apply(convert_value)
```
